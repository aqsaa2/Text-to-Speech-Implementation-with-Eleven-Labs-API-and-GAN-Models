```python
import tensorflow as tf

import yaml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as ipd

from tensorflow_tts.inference import TFAutoModel
from tensorflow_tts.inference import AutoConfig
from tensorflow_tts.inference import AutoProcessor

import librosa.display
```

```python
def visualize_mel_spectrogram(mels):
  mels = tf.reshape(mels, [-1, 80]).numpy()
  fig = plt.figure(figsize=(10, 8))
  ax1 = fig.add_subplot(311)
  ax1.set_title(f'Predicted Mel-after-Spectrogram')
  im = ax1.imshow(np.rot90(mels), aspect='auto', interpolation='none')
  fig.colorbar(mappable=im, shrink=0.65, orientation='horizontal', ax=ax1)
  plt.show()
  plt.close()

def visualize_audio_wave(wave, label=''):
  time = np.linspace(0, len(wave)/22050, len(wave))
  fig = plt.figure(figsize=(14, 8))
  ax1 = fig.add_subplot(311)
  ax2 = fig.add_subplot(311)
  ax1.plot(time, wave, 'g-', label=label)
  ax1.set(xlabel='Time (s)', ylabel='Aplitude', title='Predicted Sound Wave')
  ax1.grid()
  plt.legend()
  plt.show()
  plt.close()

  plt.figure(figsize=(14, 5))
  # librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')

def visualize_audio_waves(waves, labels):
  colors = ['b-', 'g-', 'r-', 'c-','m-','y-']
  fig = plt.figure(figsize=(14, 8))
  ax1 = fig.add_subplot(311)
  for i in range(len(waves)):
    wave = waves[i]
    color = colors[i]
    label = ''
    if len(labels) > i:
      label = labels[i]
    time = np.linspace(0, len(wave)/22050, len(wave))
    ax1.plot(time, wave, color, label=label)
  ax1.set(xlabel='Time (s)', ylabel='Aplitude', title='Predicted Sound Wave')
  # ax1.legend()
  ax1.grid()
  plt.legend()
  plt.show()
  plt.close()

  plt.figure(figsize=(14, 5))
  # librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')

```

```python
processor = AutoProcessor.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en")
```

```python
input_text = 'hello, who are you?'
input_ids = processor.text_to_sequence(input_text)
```

```python
model_stats = pd.DataFrame(columns=['Model', 'Audio Sample Length'])
```

### Tacotran 2 and Melgan Implimentation

```python
tacotron2 = TFAutoModel.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en", name="tacotron2")
melgan = TFAutoModel.from_pretrained("tensorspeech/tts-melgan-ljspeech-en", name="melgan")
```

```python
_, mel_outputs, stop_token_prediction, alignment_history = tacotron2.inference(
        tf.expand_dims(tf.convert_to_tensor(input_ids, dtype=tf.int32), 0),
        tf.convert_to_tensor([len(input_ids)], tf.int32),
        tf.convert_to_tensor([0], dtype=tf.int32)
    )
print(mel_outputs.get_shape())
audio = melgan(mel_outputs)[0, :, 0]

mels, tacotron2_melgan_audio = mel_outputs.numpy(), audio.numpy()
visualize_mel_spectrogram(mels[0])
visualize_audio_wave(tacotron2_melgan_audio)

row = ['tacotron2/melgan', len(tacotron2_melgan_audio)/22050.0]
model_stats.loc[model_stats.shape[0]] = row

ipd.Audio(tacotron2_melgan_audio, rate=22050)
```

### Fastspeech and Melgan implimentation

```python
fastspeech = TFAutoModel.from_pretrained("tensorspeech/tts-fastspeech-ljspeech-en", name="fastspeech")
```

```python
mel_before, mel_outputs, duration_outputs = fastspeech.inference(
        input_ids=tf.expand_dims(tf.convert_to_tensor(input_ids, dtype=tf.int32), 0),
        speaker_ids=tf.convert_to_tensor([0], dtype=tf.int32),
        speed_ratios=tf.convert_to_tensor([1.0], dtype=tf.float32),
    )

audio = melgan(mel_outputs)[0, :, 0]

mels, fastspeech_melgan_audio = mel_outputs.numpy(), audio.numpy()
visualize_mel_spectrogram(mels[0])
visualize_audio_wave(fastspeech_melgan_audio)

row = ['fastspeech/melgan', len(fastspeech_melgan_audio)/22050.0]
model_stats.loc[model_stats.shape[0]] = row

ipd.Audio(fastspeech_melgan_audio, rate=22050)
```
